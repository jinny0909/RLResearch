```python
from ortools.linear_solver import pywraplp 

def LinearSolver():
    solver = pywraplp.Solver.CreateSolver('SCIP')

    central_div = ['WhiteSox', 'Guardians', 'Tigers', 'Royals', 'Twins', 'Clubs', 'Reds', 'Brewers', 'Pirates', 'Cardinals']
    west_div = ['Astros', 'Angels', 'Athletics', 'Mariners', 'Rangers', 'Diamondbacks', 'Rockies', 'Dodgers', 'Padres', 'Giants']
    east_div = ['Orioles', 'Braves', 'RedSox', 'Marlins', 'Yankees', 'Mets', 'Rays', 'Phillies', 'BlueJays', 'Nationals']
    
    
    #assigning decision variables 
    all_teams = central_div + west_div + east_div 
    variables = {}
    
    for week in range(1,27): #26 * 29 * 30 
        for i in range(0,len(all_teams)): 
            for j in range(0,len(all_teams)): 
                if all_teams[i] != all_teams[j]: #not playing against itself 
                    variables[(all_teams[i], all_teams[j], week)] = solver.IntVar(0.0, 1.0, all_teams[i] + '_' + all_teams[j] + str(week))
    
    print('Total') 
    print(solver.NumVariables())
    print('number of variables')
  #  print(variables)
    
    #Constraint 1: Each team plays each of the other teams in the same divison twice
   #Central_Div 
    
#     for central_team in central_div: 
#         cent_plays = [] 
#         for opp_team in central_div: 
#             if central_team != opp_team:
#                 for week in range(1, 27): 
#                     cent_plays.append(variables[(central_team, opp_team, week)]) 
#                # print(plays)
#                 solver.Add(sum(cent_plays) == 2)
#                 cent_plays = [] 
    
#     for west_team in west_div: 
#         west_plays = [] 
#         for opp_team in west_div: 
#             if west_team != opp_team:
#                 for week in range(1, 27): 
#                     west_plays.append(variables[(west_team, opp_team, week)]) 
#                # print(plays)
#                 solver.Add(sum(west_plays) == 2)
#                 west_plays = [] 
                
#     for east_team in central_div: 
#         east_plays = [] 
#         for opp_team in east_div: 
#             if east_team != opp_team:
#                 for week in range(1, 27): 
#                     east_plays.append(variables[(east_team, opp_team, week)]) 
#                # print(plays)
#                 solver.Add(sum(east_plays) == 2)
#                 eastplays = [] 
                

## Constraint 2: Each team plays team from different division once 


## Constraint 3: Each team plays at least once a week  
 # >= 1

# playing everyday a week 
## ex) AB1 + BA1 + AC1 + CA1 + AD1 == 1
#     for week in range(1, 27): 
#         for team in all_teams: 
#             once = 0  
#             for variable in variables.keys(): 
#                 if team in variable and week in variable: 
#                     once += variables[variable]
#             solver.Add(once == 1)

# Constraint 4: Each slot is home or away 
# ex) AB1 + BA1 <= 1 
    for week in range(1,27): 
        for team in all_teams: 
            for opp_team in all_teams: 
                if team != opp_team: #not playing against each 
                    solver.Add(variables[(team, opp_team, week)] + variables[(opp_team, team, week)] <= 1)
                    

                    
## Constraint 5: Set # of home games 
    home_game = 1
    for team1 in all_teams: 
        games = [] 
        for variable in variables.keys(): 
            home = variable[0]
            if team1 == home: 
                games.append(variables[variable])
        solver.Add(sum(games) <= home_game)
        games = [] 
        
## -- end of constraints -- 
    print('Number of constraints:', solver.NumConstraints())
    
    


## objective function 
    obj = 0.0 
    
    for team in central_div: 
        for opp in central_div: 
            if team != opp: 
                idx = 1 
                for week in range(1, 27): 
                    obj += idx * variables[(team, opp, week)]
                    idx += 1 
                    
    for team in west_div: 
        for opp in west_div: 
            if team != opp: 
                idx = 1 
                for week in range(1, 27): 
                    obj += idx * variables[(team, opp, week)]
                    idx += 1 
                    
    for team in east_div: 
        for opp in east_div: 
            if team != opp: 
                idx = 1 
                for week in range(1, 27): 
                    obj += idx * variables[(team, opp, week)]
                    idx += 1 
                    
    solver.Maximize(obj)
    status = solver.Solve()
    print(status)
    if status == pywraplp.Solver.OPTIMAL: 
        print("Optimal objective functional value: ", solver.Objective().Value())
        for val in variables.values(): 
            if val.solution_value() == 1:
                print(val, val.solution_value())
    elif status == pywraplp.Solver.INFEASIBLE: 
        print('Infeasible')
    else: 
        print('Feasible, but not optimal')
    
        





                    
```
